# Chromosome Tube Visualization

```python
import pyvista as pv
import numpy as np
import mdtraj as md
```

```python
trajectory = md.load('../data/raw/chr1.pdb')
coords = trajectory.xyz[0]

surface = np.load('x_y_z_dist_surf.npy')
```

```python
# Create spline and tube with multicolor
spline = pv.Spline(coords, n_points=len(coords)*2)
tube = spline.tube(radius=0.05, n_sides=20)

# Add color gradient
scalars = np.linspace(0, 1, tube.n_points)
tube['colors'] = scalars

# Visualize with rainbow colors
plotter = pv.Plotter()
plotter.add_mesh(tube, scalars='colors', cmap='rainbow', smooth_shading=True)
plotter.add_title('Multicolor Chromosome Tube')
plotter.show()
```
