# Chromosome Tube Visualization

```python
import pyvista as pv
import numpy as np
import mdtraj as md
```

```python
trajectory = md.load('../data/raw/chr1.pdb')
coords = trajectory.xyz[0]

surface = np.load('surface_coords_scaled.npy')
```

```python
# Create spline and tube with golden color
spline = pv.Spline(coords, n_points=len(coords)*2)
tube = spline.tube(radius=0.05, n_sides=20)

# Visualize with golden color
plotter = pv.Plotter()
plotter.add_mesh(tube, color='gold', smooth_shading=True)
plotter.add_title('Golden Chromosome Tube')
plotter.show()
```

```python

surface_xyz = surface[:, :3]

# Same chromosome tube as before
spline = pv.Spline(coords, n_points=len(coords)*2)
tube = spline.tube(radius=0.05, n_sides=20)
scalars = np.linspace(0, 1, tube.n_points)
surface_points = pv.PolyData(surface_xyz)

# Plot tube + surface points
plotter = pv.Plotter(notebook=True)
plotter.add_mesh(tube, scalars=scalars, cmap='rainbow', smooth_shading=True)
plotter.add_mesh(surface_points, render_points_as_spheres=True, point_size=10, color='black', opacity=0.1)
plotter.add_title('Chromosome Tube with Surface Points')
plotter.show()
```
